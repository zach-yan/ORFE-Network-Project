```python
# 3. Create the NetworkX Graph
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# First, add all researchers from nodes_df as nodes with their attributes
for index, row in nodes_df.iterrows():
    G.add_node(row['Id'],
               Full_Name=row['Label'], # 'Label' is 'Full_Name' in df_researchers
               Institution=row['Institution'],
               Department=row['department'],
               Subject1=row['Subject 1'],
               Subject2=row['Subject 2'],
               Application1=row['Application 1'],
               Application2=row['Application 2'])

# Then, add edges from edges_df. NetworkX will automatically add nodes
# that are in edges_df but not in nodes_df (unknown collaborators).
for _, row in edges_df.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=row['Weight'])

# Remove nodes that have no edges (degree 0)
isolated_nodes = [node for node, degree in dict(G.degree()).items() if degree == 0]
G.remove_nodes_from(isolated_nodes)

# 4. Color Coding Logic
# Re-filter nodes based on the modified graph G
princeton_nodes = [node for node in nodes_df[nodes_df['Institution'] == 'Princeton University']['Id'].tolist() if node in G.nodes()]
anchor_nodes = [node for node in nodes_df[nodes_df['Institution'] != 'Princeton University']['Id'].tolist() if node in G.nodes()]
all_known = set(princeton_nodes + anchor_nodes)
unknown_nodes = [node for node in G.nodes() if node not in all_known]

# Calculate betweenness centrality for node sizing
betweenness_centrality = nx.betweenness_centrality(G)
# Scale node sizes based on betweenness centrality for better visualization
node_sizes = [v * 5000 + 100 for v in betweenness_centrality.values()] # Scale for visibility

# 5. Drawing the Graph
pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42) # k controls distance between nodes
plt.figure(figsize=(15, 12))

# Draw the 3 categories of nodes with sizes based on betweenness centrality
nx.draw_networkx_nodes(G, pos, nodelist=princeton_nodes, node_color='orange', label='Princeton (ORFE)', node_size=[node_sizes[list(G.nodes()).index(node)] for node in princeton_nodes])
nx.draw_networkx_nodes(G, pos, nodelist=anchor_nodes, node_color='skyblue', label='External Anchors (MIT/Stanford/etc)', node_size=[node_sizes[list(G.nodes()).index(node)] for node in anchor_nodes])
nx.draw_networkx_nodes(G, pos, nodelist=unknown_nodes, node_color='lightgrey', label='Unknown Collaborators', node_size=[node_sizes[list(G.nodes()).index(node)] for node in unknown_nodes])

# Draw edges, scaling the thickness by co-authorship weight
weights = [G[u][v]['weight'] for u, v in G.edges()]
# Check if there are any edges before trying to find max_w
if weights:
    max_w = max(weights)
    scaled_weights = [0.1 + (w / max_w) * 2 for w in weights]
else:
    scaled_weights = [] # No edges, so no scaled weights
nx.draw_networkx_edges(G, pos, width=scaled_weights, alpha=0.3, edge_color='gray')

# Draw text labels ONLY for known/anchored researchers to prevent visual clutter
labels = {node: node for node in princeton_nodes + anchor_nodes}
nx.draw_networkx_labels(G, pos, labels, font_size=8, font_family='sans-serif')

plt.title("Current Network Graph: ORFE Connections & External Academic Anchors (Node size by Betweenness Centrality)", fontsize=16)
plt.legend(scatterpoints=1)
plt.axis('off')
plt.tight_layout()
plt.show()

# 6. Save the final synchronized datasets
edges_df.to_csv('final_edges.csv', index=False)
nodes_df.to_csv('final_nodes.csv', index=False)
print("Graph rendered and updated CSVs successfully saved!")
```
