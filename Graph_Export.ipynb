Generate the file for visualization.

```python
import pandas as pd
import networkx as nx
from google.colab import files

# 1. Load the finalized data
# Make sure these filenames match exactly what is in your Colab files
nodes_df = pd.read_csv("orfe_nodes_final_gravity.csv")
edges_df = pd.read_csv("orfe_edges_fuzzy_cleaned.csv")

# 2. Clean data types for GEXF compatibility
# NetworkX will fail if it encounters NaNs or Numpy data types during export
nodes_df.fillna("Unknown", inplace=True)

# 3. Initialize the Graph
G = nx.Graph()

print("Adding nodes to the graph...")
for _, row in nodes_df.iterrows():
    # Force everything into standard Python types (str, int, float)
    node_id = str(row['Id'])

    G.add_node(
        node_id,
        label=str(row['Label']),
        layer=float(row.get('Layer', 2.0)),
        affiliation=str(row['Affiliation']),
        # Keep lists as strings so Gephi can import them as text attributes
        subjects=str(row.get('Subjects', '[]')),
        applications=str(row.get('Applications', '[]'))
    )

print("Adding edges to the graph...")
for _, row in edges_df.iterrows():
    source = str(row['Source'])
    target = str(row['Target'])

    # Ensure edges only connect existing nodes
    if G.has_node(source) and G.has_node(target):
        # Force weight to a standard Python float
        weight = float(row.get('Weight', 1.0))
        G.add_edge(source, target, weight=weight)

# 4. Output Basic Graph Metrics
print("\n--- Network Summary ---")
print(f"Total Nodes: {G.number_of_nodes():,}")
print(f"Total Edges: {G.number_of_edges():,}")

# 5. Export for Gephi
export_filename = "orfe_coauthorship_network.gexf"
nx.write_gexf(G, export_filename)
print(f"\nSuccessfully generated {export_filename}!")

# 6. Download the file
files.download(export_filename)
```
